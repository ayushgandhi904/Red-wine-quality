```python
#Feature columns:
x = ["fixed acidity", "volatile acidity", "citric acid", "residual sugar", "chlorides", "free sulfur dioxide", "total sulfur dioxide", "density", "pH", "sulphates", "alcohol"]
y = ["quality"]
```

```python
#Histplot

for i in x:
    plt.figure(figsize=(3,3))
    plt.tight_layout()
    plt.hist(df[i], bins = 10)
    plt.title(f"{i} hist plot")
    plt.show()
```

```python
#scatter plot

for i in x:
    plt.figure(figsize=(4, 4))
    plt.scatter(df[i], df["quality"])
    plt.title(f"{i} scatter plot")
    plt.show()
```

```python
#Heatmap

plt.figure(figsize=(15, 15))
sns.heatmap(df.corr(), annot = True, cmap="Greens")
plt.title("Heatmap")
plt.show()
```

```python
#Pairplot
sns.set_palette("husl")
sns.pairplot(df)
plt.title("Pair plot")
plt.show()
```
